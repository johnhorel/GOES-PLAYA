First, get the HRRR lat/lon grid

Open a GLM file

```python
sDATE = datetime(2018, 5, 1)
eDATE = datetime(2018, 5, 2)
HOURS = [0, 1, 2]
FILES = get_GLM_files_for_range(sDATE, eDATE, HOURS=HOURS)
```

```python
len(FILES['Files'])
```

```python
GLM = accumulate_GLM_FAST(FILES)
```

```python
m.scatter(GLM['longitude'], GLM['latitude'], marker='+', latlon=True, s=1)
m.drawcountries()
m.drawcoastlines()
plt.title('Raw GLM Flash Location', loc='left', fontweight='semibold')
plt.title('Start: %s\n End: %s' % (sDATE, eDATE), loc='right')
```

```python
bins = (800, 895)
H, y, x = np.histogram2d(GLM['longitude'], GLM['latitude'], bins=bins)
Y, X, = np.meshgrid(y,x)
H = np.transpose(H)
```

```python
masked = H
masked = np.ma.array(masked)
masked[masked == 0] = np.ma.masked
```

## Map: GLM Field of View

```python
B.pcolormesh(Y, X, H, latlon=True, cmap='magma', vmin=0, vmax=np.percentile(H, 99.99))
cb = plt.colorbar(orientation='horizontal', pad=.01, shrink=.6)
cb.set_label(r'GLM Flash Count (~14 km$\mathregular{^{2}}$ bins)')

B.drawcoastlines(color='grey')
B.drawcountries(color='grey')

plt.title('GOES-16 GLM Flashes', loc='left', fontweight='semibold')
plt.title('Start: %s\n End: %s' % (sDATE.strftime('%d %b %Y'), eDATE.strftime('%d %b %Y')), loc='right', fontsize=10)
plt.ylabel(r'~14 km$\mathregular{^{2}}$ bins')

#plt.savefig('GLM_field-of-view')
```

## Map: HRRR CONUS Domain

```python
m.pcolormesh(Y, X, H, latlon=True, vmin=0, vmax=np.percentile(H, 99.99), cmap='magma')
cb = plt.colorbar(orientation='horizontal', pad=.01, shrink=.8, extend='max')
cb.set_label(r'GLM Flash Count (~14 km$\mathregular{^{2}}$ bins)')

m.drawcoastlines(color='grey')
m.drawcountries(color='grey')
m.drawstates(color='grey')

plt.title('GOES-16 GLM Flashes', loc='left', fontweight='semibold')
plt.title('Start: %s\n End: %s' % (sDATE.strftime('%d %b %Y'), eDATE.strftime('%d %b %Y')), loc='right', fontsize=10)
plt.ylabel(r'~14 km$\mathregular{^{2}}$ bins')

#plt.savefig('GLM_CONUS')
```

## Map: Western United States

```python
mU.pcolormesh(Y, X, H, latlon=True, vmin=0, vmax=np.percentile(H, 99.9), cmap='magma')
cb = plt.colorbar(orientation='horizontal', pad=.01, shrink=.6, extend='max')
cb.set_label(r'GLM Flash Count (~14 km$\mathregular{^{2}}$ bins)')

mU.drawcoastlines(color='grey')
mU.drawcountries(color='grey')
mU.drawstates(color='grey')
mU.drawcounties(color='grey')

plt.title('GOES-16 GLM Flashes', loc='left', fontweight='semibold')
plt.title('Start: %s\n End: %s' % (sDATE.strftime('%d %b %Y'), eDATE.strftime('%d %b %Y')), loc='right', fontsize=10)
plt.ylabel(r'~14 km$\mathregular{^{2}}$ bins')

#plt.savefig('GLM_West')
```

## Size of Grid Box Bins

```python
from metpy.calc import lat_lon_grid_deltas
from metpy.units import units

yy = np.linspace(y.min(), y.max(), bins[0]+1)
xx = np.linspace(x.min(), x.max(), bins[1]+1)

YY, XX = np.meshgrid(yy, xx)
y_m, x_m = lat_lon_grid_deltas(YY,XX)

y_km = y_m.to(units.km)
x_km = x_m.to(units.km)

print('Average Grid Length\n   y length: %s\n   x length: %s' % (np.mean(y_km), np.mean(x_km)))

fig, (ax1, ax2) = plt.subplots(1,2)

plt.sca(ax1)
B.pcolormesh(YY, XX, np.round(y_km, 2), latlon=True)
cb = plt.colorbar(orientation='horizontal', pad=.01, shrink=.8)
cb.set_label(r'Longitude Grid Length (km)')
B.drawcoastlines()

plt.sca(ax2)
B.pcolormesh(YY, XX, np.round(x_km, 2), latlon=True)
cb = plt.colorbar(orientation='horizontal', pad=.01, shrink=.8)
cb.set_label(r'Latitude Grid Length (km)')
B.drawcoastlines()

#print(np.round(x_m/1000, 2))
```

## GLM Field-of-view (Flash Mask)

```python
B.pcolormesh(Y, X, H>0, latlon=True)
cb.set_label('GLM Flash Count')

B.drawcoastlines(color='grey')
B.drawcountries(color='grey')
B.drawstates(color='grey')

plt.title('GOES-16 GLM Flashes', loc='left', fontweight='semibold')
plt.title('Start: %s\n End: %s' % (sDATE.strftime('%d %b %Y'), eDATE.strftime('%d %b %Y')), loc='right', fontsize=10)
plt.ylabel(r'~14 km$\mathregular{^{2}}$ bins')

#plt.savefig('GLM_field-of-view_MASK')
```
